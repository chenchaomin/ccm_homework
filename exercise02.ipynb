```python
even = 0
count = 1
while count<100:
    count+=1
    if count % 2 == 0:
        print("2-100的偶数",count)
        even += count 
    else:
        print("2-100的奇数",count)
        even -= count 
print(even) 
```

```python
uneven = 0  
for i in range(1,100):
    if i % 2 == 1:
        print("1-99的奇数",i)
        uneven += i
    else:
        print("1-99的偶数",i)
        uneven -= i
print(uneven)  
```
